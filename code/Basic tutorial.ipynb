```python
# Line graph
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize your graph
plt.figure(figsize = (5, 3), dpi = 200)

## Line number one

# plt.plot(x, y, label = '2x', color = 'red', linewidth = 2, linestyle = '--', marker = '.', markersize = 10, markeredgecolor = 'blue')
plt.plot(x, y, 'r^--', label = '2x')

## Line number two

x2 = np.arange(0, 4.5, 0.5)

#plt.plot(x2, x2**2, label = 'x^2')
plt.plot(x2[:5], x2[:5]**2, 'b', label = 'x^2')
plt.plot(x2[4:], x2[4:]**2, 'b--')

plt.title('Our first graph', fontdict = {'fontname' : 'MS Reference Sans Serif', 'fontsize': 16})
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')

#plt.xticks([1, 2, 3, 4])
#plt.yticks([1, 3, 5, 7, 9])

plt.legend()

#plt.savefig('mygraph.png', dpi = 300)

plt.show()
```

```python
# Bar Chart
labels = ['A', 'B', 'C']
values = [2, 5, 9]

bars = plt.bar(labels, values)

patterns = ['/', '.', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('.')
# bars[2].set_hatch('*')

plt.figure(figsize = (6, 4))

plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(50, 8, 150)  #np.random.normal(mean, standard deviation, size) is a func that specifically generates random numbers from a normal distribution
#print(data)

plt.hist(data, bins = 5, range = (40, 60), color = '#abcdef', edgecolor = 'b')

plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
```

```python
# Pie Chart

import matplotlib.pyplot as plt
import numpy as np

y = np.array([36, 24, 18, 8, 14])

labels = ['serenity', 'curiosity', 'excitement', 'joy', 'contentment']
explode = [0.2, 0, 0, 0, 0]
colors = ['c', 'm', 'b', 'k', 'g']  # 'c' - Cyan  'm' - Magenta

plt.pie(y, labels = labels, startangle = 90 , explode = explode, shadow = True, colors = colors)

plt.legend(title = 'My emotions')
plt.show()
```

```python
# Boxplot(Whisker Plot)

import numpy as np
import matplotlib.pyplot as plt

# Creating dataset
np.random.seed(6)
data1 = np.random.normal(100, 20, 200)
data2 = np.random.normal(70, 40, 200)

plt.figure(figsize = (10, 7))

plt.boxplot([data1, data2])

plt.show()
```
